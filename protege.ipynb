```
# example Athene

from Athene.reasoner.knowledgebase.axioms import And,Or,Not,ClassAssertion,ABoxAxiom,RoleAssertion,TBoxAxiom,Subsumption
from Athene.reasoner.common.constructors import Concept,All,Some,Instance



little_kb=[
  ABoxAxiom(ClassAssertion(Concept("Man"),Instance("Aditya"))),
  TBoxAxiom(Subsumption(Concept("Man"),Concept("Biological")))
]

print(little_kb)
```

```
# Athene and ontology

import re
```

```
# Concepts

elems=root.findall('{http://www.w3.org/2002/07/owl#}Class')

zebras, horses, bears = '', '', ''
mammals,reptilies = '',''

for elem in elems:
  attrib = elem.attrib.get('{http://www.w3.org/1999/02/22-rdf-syntax-ns#}about')
  splitted = attrib.split('#')[1]

  if splitted.find('Челов')>=0:
    human = splitted

  if splitted.find('Доярк')>=0:
    doyark = splitted
  
  if splitted.find('Ребен')>=0:
    childs = splitted

  if splitted.find('Женщ')>=0:
    woman = splitted

  if splitted.find('Двор')>=0:
    cleaner = splitted
    
  if splitted.find('Соб')>=0:
    dogs = splitted


print(TBoxAxiom(Subsumption(Concept(doyark),Concept(woman))))
print(TBoxAxiom(Subsumption(And(Concept(woman),Concept(doyark)),Concept(human))))
print(TBoxAxiom(Subsumption(Concept(cleaner),Not(Concept(childs)))))
print()
```

```
# Assertions

instances=root.findall('{http://www.w3.org/2002/07/owl#}NamedIndividual')

zebraInstance, horseInstance, bearInstance = '', '', ''
zebraInstances, horseInstances, bearInstances = [],[],[]

mammalArray = []

for elem in instances:
  attrib = elem.attrib.get('{http://www.w3.org/1999/02/22-rdf-syntax-ns#}about')
  splitted = attrib.split('#')[1]

  mammalArray.append(splitted)

  if splitted.find('Алиш')>=0:
    bearInstance = splitted
    bearInstances.append(bearInstance)

    print(ABoxAxiom(ClassAssertion(Concept(human),Instance(bearInstance))))
    print(ABoxAxiom(ClassAssertion(Concept(childs),Instance(bearInstance))))

  if splitted.find('З')>=0:
    zebraInstance = splitted
    zebraInstances.append(zebraInstance)

    print(ABoxAxiom(ClassAssertion(Concept(doyark),Instance(zebraInstance))))

  if splitted.find('Анто')>=0:
    horseInstance = splitted
    horseInstances.append(horseInstance)

    print(ABoxAxiom(ClassAssertion(Concept(human),Instance(horseInstance))))

print()
```

```
# ObjectProperty

objects=root.findall('{http://www.w3.org/2002/07/owl#}ObjectProperty')

for elem in objects:
  attrib = elem.attrib.get('{http://www.w3.org/1999/02/22-rdf-syntax-ns#}about')
  splitted = attrib.split('#')[1]
  # print(splitted)


  if splitted.find('корм')>=0:
    print(TBoxAxiom(Subsumption(And(Concept(dogs),Concept(cleaner)),Concept(splitted))))
  
```
